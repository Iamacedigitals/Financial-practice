```python
import pandas as pd
import numpy as np
```

```python
data =  pd.read_csv("BTC.csv")
data["unix"] = pd.to_datetime(data["unix"], unit = "s")
```

```python
df1 = data.set_index("unix")
```

```python
df1 = df1.resample("24h").agg({
    "open": "first",
    "high": "max",
    "low": "min",
    "close": "last",
    "Volume USD": "sum"
})
```

```python
def basic_returns(previous_price, current_price):
    returns = (current_price - previous_price)/previous_price 
    # This calculates the backward reutrns backward returns measures the increase from the previous price to new price
    # forward returns =  (previous_price - current_price)/current_price,  will be negative if current price is higher, measures by what amount is the previous price less than the current price
    return returns
def log_returns(previous_price, current_price):
    import numpy as np
    returns = np.log(current_price/previous_price)
    return returns
```

```python
returns = pd.DataFrame({
    "Basic Returns X":basic_returns(df1["close"].shift(), df1["close"]), 
    "Log Returns X":log_returns(df1["close"].shift(), df1["close"])
}) 
```

```python
cummulative_returns = returns.cumsum()
# np.sum(returns)
```
